```python
import pandas as pd
import matplotlib.pyplot as plt

students = pd.DataFrame({
    "Name": ["Arai", "West", "Masha", "Liza", "Adilya"],
    "Score": [90, 95, 88, 75, 80]
})

students
```

```python
plt.plot(students["Name"], students["Score"])
plt.xlabel("Students")
plt.ylabel("Score")
plt.title("Student Scores Line Plot")
plt.show()
```

```python
plt.bar(students["Name"], students["Score"])
plt.xlabel("Students")
plt.ylabel("Score")
plt.title("Student Scores Bar Chart")
plt.show()
```
